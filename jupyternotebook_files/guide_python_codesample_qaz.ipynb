## 3-бөлім. Бағдарламаның миы: Шарттар мен циклдер

Шарттар (`if`) кодты тек белгілі бір шарт орындалғанда ғана іске қосуға мүмкіндік береді. Циклдер (`for`) кодты бірнеше рет қайталауға мүмкіндік береді. **НАЗАР АУДАРЫҢЫЗ:** Python-да жол басындағы бос орындар (шегіністер) өте маңызды! Олар қай кодтың шарттың немесе циклдің 'ішінде' екенін анықтайды.

```python
# `if` шартының мысалы:
m = 0 # Бастапқы санауыш
# Кубикті бір рет лақтыруды модельдейік
roll = 6 

# Алтының түскенін тексерейік (== 'тең бе?' дегенді білдіреді)
if roll == 6:
    print("Сәтті болды! Алты түсті.")
    m = m + 1 # Санауышты арттырамыз

print("Санауыштың жаңа мәні:", m)
```

```python
# `for` циклының мысалы:
# Әрекетті 5 рет қайталайық
# range(5) 0-ден 4-ке дейінгі сандар тізбегін жасайды
for i in range(5):
    # Төмендегі шегінісі бар код 5 рет орындалады
    print("Бұл қайталану нөмірі", i)
```

## 4-бөлім. Құралдарды қосу: Кітапханалар

Кітапханалар — бұл дайын құралдар жиынтығы. Оларды пайдалану үшін бағдарламаның басында 'импорттау' қажет. Біз үш негізгі кітапхананы қолданамыз.

```python
# Кітапханаларды импорттау (ноутбуктың басында бір рет жасалады):
import random                # Кездейсоқ сандарды генерациялау үшін
import numpy as np           # Ыңғайлы математикалық есептеулер үшін
import matplotlib.pyplot as plt # Графиктерді салу үшін
```

_`random.randint(a, b)`: `a`-дан `b`-ға дейінгі аралықта кездейсоқ БҮТІН сан генерациялайды (кубик үшін)._

```python
dice_roll = random.randint(1, 6)
print("Кубикте түскен сан:", dice_roll)
```

_`random.uniform(a, b)`: `a`-дан `b`-ға дейінгі аралықта кездейсоқ НАҚТЫ сан генерациялайды (уақыт, координаталар үшін)._

```python
time = random.uniform(0, 60)
print("Кездейсоқ уақыт мезеті:", time)
```

_`np.mean(сандар_тізімі)`: сандар тізімінің орташа мәнін лезде есептейді._

```python
results = [5, 2, 6, 1, 3, 5]
average_roll = np.mean(results)
print("Лақтырулардың орташа мәні:", average_roll)
```

_Қарапайым график салу мысалы:_

```python
x_axis_data = [20, 100, 500, 1000] # Сынақтар саны
y_axis_data = [0.0333, 0.0167, 0.0067, 0.0007] # Ауытқу |W - P|

plt.plot(x_axis_data, y_axis_data, marker='o') # Графикті құру (нүктелері бар сызық)
plt.title("Жиіліктің жинақталуы")   # Тақырып қосу
plt.xlabel("Сынақтар саны (n)") # X осінің жазуы
plt.ylabel("Ауытқу |W - P|")  # Y осінің жазуы
plt.grid(True)                     # Торды қосу
plt.show()                         # Графикті көрсету
```

## 5-бөлім. Барлығын біріктіру: Шешім үлгісі (шаблон)

Енді қарапайым есепті басынан аяғына дейін шешіп көрейік. Бұл код — зертханалық жұмыстың тапсырмаларын орындауға арналған негізгі үлгіңіз.

**Есеп:** 1000 рет тиын лақтыруды модельдеу. 'Елтаңба' = 1, 'Сан' = 0. 'Елтаңба' қанша рет түскенін санап, салыстырмалы жиілікті табу.

```python
# 1-қадам: Қажетті құралдарды импорттау (егер бұрын жасамасаңыз)
import random

# 2-қадам: Бастапқы шарттарды белгілеу
n = 1000  # Жалпы лақтырулар саны
m = 0     # 'Елтаңба' түсу санауышы

# 3-қадам: Эксперимент циклын іске қосу
for i in range(n):
    # Цикл ішінде бір лақтыруды модельдейміз
    coin_flip = random.randint(0, 1) # 0 - сан, 1 - елтаңба
    
    # 4-қадам: Шартты тексеру
    if coin_flip == 1:
        # Егер елтаңба түссе, санауышты арттырамыз
        m = m + 1

# 5-қадам: Нәтижелерді өңдеу және шығару
# Цикл аяқталды, енді бізде соңғы m мәні бар
print("Жалпы лақтырулар саны (n):", n)
print("'Елтаңба' түсу саны (m):", m)

# Салыстырмалы жиілікті есептеу
relative_frequency = m / n
print("Салыстырмалы жиілік:", relative_frequency)
```

**Қорытынды:** Зертханалық жұмыстағы барлық тапсырмалар дәл осындай логикамен құрастырылған: **Импорттау -> Инициализация -> Цикл -> Цикл ішіндегі шарт -> Қорытынды есептеу**.
